### Answering business questions by visualizing our data

### Question 1 : What was the best month for sales? How much was earned that month?

To answer this question we will be using the sales column from our data set and will plot it against months.

```python
Data_Sales = Data.groupby('Month').sum()
Data_Sales['Sales'] = Data_Sales['Sales']/10
Data_Sales['Sales'] = Data_Sales['Sales'].round(1)
```

```python
months = range(1,13)

x = months
y = Data_Sales['Sales']

plt.rcParams['figure.figsize'] = (13, 7)
plt.style.use('fivethirtyeight')

plt.bar(x , y , color = 'chocolate')
plt.title("Sales Per Month" , size = 25)
plt.xticks(months)
plt.xlabel("Months" , size = 15)
plt.ylabel("Sales($)" , size = 15)

plt.show()
```

So, as the answer to our question we got that the maximum sales has been observed in the month of "December". The reason might be "Christmas" , but we cannot claim anything right now as their might be other internal factors like marketing campaign or else affecting the sales in the month of december. We can look for that later.

Similar can be done to determine the reason for getting the minimum sales in "January".

### Question 2 : What city has the highest number of sales?

To answer this question we will be using the city coulmn and the sales column and will plot them.

```python
Data_Sales = Data.groupby('City').sum()
Data_Sales['Sales'] = Data_Sales['Sales']/10
Data_Sales['Sales'] = Data_Sales['Sales'].round(1)
```

```python
x = [city for city , df in Data.groupby('City')]
y = Data_Sales['Sales']

plt.rcParams['figure.figsize'] = (13, 7)
plt.style.use('fivethirtyeight')

plt.bar(x , y , color = 'crimson')
plt.title("Sales per City " , size = 30)
plt.xticks(x , rotation = 'vertical' , size = 12)
plt.xlabel("City" , size = 20)
plt.ylabel("Sales($)" , size = 20)

plt.show()
```

As the answer to the second question we get that the maximum number of sales is observed in "San Francisco CA". There might be several reasins for this result. ### Question 4 : What products are most often sold together?

```python
Data.head()
```

```python
df = Data[Data['Order ID'].duplicated(keep = False)]
df.head(20)
```

Okay. So we have got the list of all the products bought together. But to answer the question and create a visualization we have to merge it in a proper/usable manner.

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x : ','.join(x))
df.head()
```

Okay. But now in this data frame, we have duplicate entries. SO, let's just drop the duplicates to get the data we want to create a visualization.

```python
df = df[['Order ID' , 'Grouped']].drop_duplicates()
df.head()
```

```python
#Referenced from Stackoverflow
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list , 2)))
    
df_1 = count.most_common(10)
    
for key, value in count.most_common(10):
    print(key , value)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list , 3)))
    
df_2 = count.most_common(10)
    
for key, value in count.most_common(10):
    print(key , value)
```

### Question 5 : What Product sold the most? Why do you think it sold the most?

```python
product_group = Data.groupby('Product')
product_group.sum()
```

```python
product_group = Data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product , df in product_group]

x = products
y = quantity_ordered

plt.rcParams['figure.figsize'] = (13, 7)
plt.style.use('fivethirtyeight')

plt.bar(x , y , color = 'navy')
plt.title("Products Ordered Most " , size = 30)
plt.xticks(x , rotation = 'vertical' , size = 12)
plt.xlabel("Products" , size = 20)
plt.ylabel("Quantity Ordered" , size = 20)

plt.show()
sb.set()
```

Okay. So we have got the visualization for the products sold most or least or the number of products sold. From the graph we can say that "AAA Batteries(4-pack)" is sold the most and "LG Dryer" and "LG Washing Machine" are the ones sold least. The reason might be the lower price for batteries and expensive Washine Machine. Let's visualize that as well.

```python
prices = Data.groupby('Product').mean()['Price Each']
prices
```

```python
f , ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(x, y , color = 'navy')
ax1.set_xlabel('Product Name' , )
ax1.set_ylabel('Quantity Ordered')
ax1.set_xticklabels(x , rotation = 90 , size = 12)

ax2.plot(x , prices , 'b-' , color = 'goldenrod' , linewidth = 4)
ax2.set_ylabel('Price ($)' , color = 'black')

plt.title("Products Ordered Most vs Prices " , size = 30)

ax2.grid(False)

plt.show()
sb.set()
```

Okay, Great!!
